```python
import numpy as np

sigma = 1.4
mask = np.zeros((7, 7))

for x in range(-3, 4):
    for y in range(-3, 4):
        mask[x + 3, y + 3] = np.round(np.exp((-x * x - y * y) / (2 * sigma**2)), 4)

print(mask)
```

```python
min_value = np.min(mask)
normalized_mask = mask / min_value
normalized_mask_int = np.round(normalized_mask,0).astype(np.uint8)

print(normalized_mask_int)
print("1/K^2: ", "1/", np.sum(normalized_mask_int))
```
